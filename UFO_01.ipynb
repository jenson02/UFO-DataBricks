#Clean the column names
This makes the dataset easier to work with by renaming columns that contain spaces or trailing spaces.

```python
df_sightings_clean = (
    df_sightings
    .withColumnRenamed("duration (seconds)", "duration_seconds")
    .withColumnRenamed("duration (hours/min)", "duration_hours_min")
    .withColumnRenamed("date posted", "date_posted")
    .withColumnRenamed("longitude ", "longitude")
)
 
display(df_sightings_clean)
```

#Check the cleaned columns

```python
df_sightings_clean.printSchema()
```

#Count sightings by state

```python
state_counts = (
    df_sightings_clean
    .groupBy("state")
    .count()
    .orderBy("count", ascending=False)
)
 
display(state_counts)
```

#Compare the columns in both datasets
Before combining datasets, it is important to check what columns each dataset contains.

```python
print("UFO Sightings columns:")
print(df_sightings_clean.columns)
 
print("\nUFO Truth columns:")
print(df_truth.columns)
```

#Join the two datasets
The two datasets do not share a city column, so we will combine them using shape column.

This allows us to merge UFO sightings with additional investigation information such as weather witness, and research outcome.

#**Inner join = only rows where the UFO shape appears in both datasets**

```python
combined_df = df_sightings_clean.join(
    df_truth,
    on="shape",
    how="inner"
)
 
display(combined_df)
```

#Visualisation: UFO sightings by shape
Now that the datasets are combined we can count how many sightings exsist for each UFO shape.

```python
shape_mention_counts = (
    combined_df
    .groupBy("shape")
    .count()
    .orderBy("count", ascending=False)
)

display(shape_mention_counts)
```

#Visualisation: Research outcomes by UFO shape
The investigation dataset contains a column called researchOutcome. This shows what investigators concluded about the sighting.

```python
outcome_by_shape = (
    combined_df
    .groupBy("shape", "researchOutcome")
    .count()
    .orderBy("count", ascending=False)
)
 
display(outcome_by_shape)
```

#Visualisation:Physical evidence reported
This chart shows whether physical evidence was reported fro each UFO shape.

```python
evidence_by_shape = (
    combined_df
    .groupBy("shape", "physicalEvidence")
    .count()
    .orderBy("count", ascending=False)
)
 
display(evidence_by_shape)
```

#Visualisation:Weather condistions during sightings

```python
weather_by_shape = (
    combined_df
    .groupBy("shape", "weather")
    .count()
    .orderBy("count", ascending=False)
)
 
display(weather_by_shape)
```

#Visualisation: Average number of witnesses by UFO shape
This chart shows which UFO Shape tend to have the highest amount number of witnesses.

```python
from pyspark.sql.functions import avg, col
from pyspark.sql.types import IntegerType
 
combined_witnesses = combined_df.withColumn(
    "witnesses_num",
    col("witnesses").cast(IntegerType())
)
 
avg_witnesses = (
    combined_witnesses
    .groupBy("shape")
    .agg(avg("witnesses_num").alias("avg_witnesses"))
    .orderBy("avg_witnesses", ascending=False)
)
 
display(avg_witnesses)
```

#Visualisation: Number of UFO sightings by state
This chart shows which states have the most sightings.

```python
state_counts = (
    df_sightings_clean
    .groupBy("state")
    .count()
    .orderBy("count", ascending=False)
)
 
display(state_counts)
```

#Visualisation: Map of UFO sightings
This map uses latitude and longitude to plot sightings geographically.

#Prepare Coordinates for mapping
Latitude and longitude must be numeric values for the map visualisation to work.

```python
from pyspark.sql.functions import col
from pyspark.sql.types import DoubleType
 
map_df = (
    df_sightings_clean
    .withColumn("latitude", col("latitude").cast(DoubleType()))
    .withColumn("longitude", col("longitude").cast(DoubleType()))
)
 
display(map_df)
```

#Remove rows with missing coordinates

```python
map_sample = map_df.select(
    "latitude",
    "longitude",
    "shape",
    "country"
).limit(5000)
 
display(map_sample)
```
